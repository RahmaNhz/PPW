# Part 3. PENERAPAN MODEL LOGISTIC REGRESSION UNTUK KLASIFIKASI BERITA


Logistic Regression merupakan metode dalam pembelajaran mesin yang termasuk dalam supervised learning dan digunakan untuk memprediksi probabilitas hasil biner 1/0 (true/false) berdasarkan satu atau beberapa variabel fitur, degan menggunakan fungsi logistic (sigmoid) untuk melakukan prediksi.

**Rumus Fungsi Sigmoid:**
$$
y = f(x) = \frac{1}{1 + e^{-z}}
$$

Di mana \( z \) adalah kombinasi linier dari variabel input (prediktor):

$$
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n
$$

Maka:

$$
\hat{y} = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n)}}
$$

### Hasil Preprosessing Berita

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/PPW/tugas/preprocessing-cnnnews.csv")
df.head()
```

###TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Menginisialisasi TfidfVectorizer
vectorizer = TfidfVectorizer()
# Menghitung TF-IDF
tfidf_matrix = vectorizer.fit_transform(df['stopword_removal'])
```

```python
# Mengubah hasilnya menjadi DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head(10)
```

### MEMBANGUN MODEL LOGISTIC REGGRESION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# kolom yang berisi label target
y = df['kategori']
```

```python
# Membagi data menjadi data latih dan uji (80% latih, 20% uji)
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, y, test_size=0.2, random_state=42)
```

```python
# Membuat model Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Membuat prediksi pada data uji
y_pred = logreg.predict(X_test)
```

```python
# Mengevaluasi model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import pickle

# Simpan TfidfVectorizer
with open('tfidf_vectorizer.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)
```

```python
# Simpan model Logistic Regression ke file pickle
with open('model_logreg.pkl', 'wb') as f:
    pickle.dump(logreg, f)

print("Model Logistic Regression telah disimpan ke file 'model_logreg.pkl'")
```
