## Import data berita yang telah di preprosessing

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/PPW/tugas/preprocessing-cnnnews.csv")
df.head()
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Menginisialisasi TfidfVectorizer
vectorizer = TfidfVectorizer()
# Menghitung TF-IDF
tfidf_matrix = vectorizer.fit_transform(df['stopword_removal'])
```

```python
# Mengubah hasilnya menjadi DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

tfidf_df.to_csv("tfidf.csv", encoding='utf-8', index=False)

tfidf_df.head(10)
```

## Cosine similarity

Cosine similarity adalah ukuran kesamaan antara dua vektor dalam ruang berdimensi tinggi. Nilai cosine similarity berkisar antara -1 dan 1, di mana 1 berarti vektor tersebut identik, 0 berarti tidak ada kesamaan, dan -1 berarti vektor tersebut berlawanan secara sempurna.  Semakin kecil sudutnya, semakin besar nilai cosine similarity, yang berarti kedua vektor tersebut lebih mirip.

Rumus Cosine Similarity:

Jika kita memiliki dua vektor ( A ) dan ( B ), cosine similarity dapat dihitung dengan rumus berikut:

$$ \text{Cosine Similarity} = \cos(\theta) = \frac{A \cdot B}{|A| |B|} $$

Di mana:

$( A \cdot B )$ adalah hasil perkalian dot $(dot product)$ dari vektor ( A ) dan ( B ).
$( |A| )$ adalah norma (panjang) dari vektor ( A ).
$( |B| )$ adalah norma (panjang) dari vektor ( B ).



```python
from sklearn.metrics.pairwise import cosine_similarity

# Menghitung cosine similarity berdasarkan matriks TF-IDF dari kolom 'isi'
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=df.index, columns=df.index)

# Tampilkan beberapa baris dari cosine similarity matrix
cosine_sim_df.head(100)
```

## Konversi

```python
# Mengonversi cosine similarity menjadi 0 dan 1
binary_cosine_sim_matrix = (cosine_sim_matrix >= 0.5).astype(int)

# Mengubah hasilnya menjadi DataFrame
binary_cosine_sim_df = pd.DataFrame(binary_cosine_sim_matrix, index=df.index, columns=df.index)

binary_cosine_sim_df.head(100)
```

# Graf

Graf adalah sebuah himpunan objek-objek yang dinamakan titik, simpul, atau sudut yang dihubungkan oleh penghubung yang dinamakan garis atau sisi. Graf digunakan untuk merepresentasikan hubungan antara objek-objek.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf dari matriks biner
G = nx.from_pandas_adjacency(binary_cosine_sim_df)

# Menggambar graf
plt.figure(figsize=(12, 12))  # Menentukan ukuran figure
G.remove_edges_from([(u, u) for u in G.nodes()])
pos = nx.random_layout(G) # Layout untuk visualisasi graf
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700,
        font_size=15, font_color='black', edge_color='gray')



# Menampilkan graf
plt.title("Graf Keterhubungan Berita Berdasarkan Cosine Similarity")
plt.show()
```

### Centrality dari dokumen Berita

```python
import networkx as nx
import pandas as pd

# Membuat Graph dari matriks biner cosine similarity
G = nx.Graph(binary_cosine_sim_df.values)

# Menghitung Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Menghitung Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)

#Menghitung Closeness Centrality
closeness_centrality = nx.betweenness_centrality(G)

# Mengonversi hasil ke DataFrame
degree_centrality_df = pd.DataFrame(list(degree_centrality.items()), columns=['Index', 'Degree Centrality'])
betweenness_centrality_df = pd.DataFrame(list(betweenness_centrality.items()), columns=['Index', 'Betweenness Centrality'])
closeness_centrality_df = pd.DataFrame(list(closeness_centrality.items()), columns=['Index', 'Closeness Centrality'])

# Menampilkan beberapa hasil
print("Degree Centrality:")
print(degree_centrality_df.head())

print("Betweenness Centrality:")
print(betweenness_centrality_df.head())

print("Closeness Centrality:")
print(closeness_centrality_df.head())

# Menemukan berita dengan Degree Centrality tertinggi
top_degree = betweenness_centrality_df.nlargest(10, 'Betweenness Centrality')
print("\nTop 10 Berita dengan Betweenness Centrality Tertinggi:")
print(top_degree)
```
